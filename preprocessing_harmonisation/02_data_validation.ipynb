```python
# ===================================================================
# CELL 1: VISUALLY COMPARE SPECTROGRAMS FROM DIFFERENT SOURCES
# ===================================================================
import pandas as pd
from pathlib import Path
import os
import matplotlib.pyplot as plt
from PIL import Image

# --- CONFIGURATION ---
# This function robustly finds your project's root directory
def find_project_root(start_path, marker='README.md'):
    path = Path(start_path).resolve()
    while not (path / marker).exists():
        if path.parent == path:
            raise FileNotFoundError(f"Could not find project root with marker '{marker}'")
        path = path.parent
    return path

try:
    PROJECT_ROOT = find_project_root(os.getcwd())
    print(f"✅ Successfully found Project Root at: {PROJECT_ROOT}")
    
    # This path is now updated to your specific training run
    METADATA_CSV = PROJECT_ROOT / "training_runs/run_20250808_183050_manual_lora/metadata.csv" 

    # --- Load Data ---
    metadata = pd.read_csv(METADATA_CSV)
    print(f"✅ Full dataset loaded with {len(metadata)} samples.")

    # --- VISUALIZATION ---
    # Get a list of the unique sources in your data
    sources = metadata['source'].unique()
    
    # Create a plot with one subplot for each source
    fig, axes = plt.subplots(1, len(sources), figsize=(20, 5))
    fig.suptitle('Visual Comparison of Spectrograms by Source', fontsize=16)

    for i, source in enumerate(sources):
        # Get one random sample from the current source
        sample = metadata[metadata['source'] == source].sample(n=1, random_state=42).iloc[0]
        img_path = Path(sample['filepath'])
        
        try:
            img = Image.open(img_path)
            ax = axes[i]
            ax.imshow(img)
            ax.set_title(f"Source: {source}\nLabel: {sample['label']}")
            ax.axis('off')
        except FileNotFoundError:
            print(f"File not found for source '{source}': {img_path}")
            
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    
except FileNotFoundError:
    print(f"❌ [ERROR] Could not find the metadata CSV at: {METADATA_CSV}")
    print("Please update the path to the 'metadata.csv' from your latest PEFT training run.")
except Exception as e:
    print(f"❌ [ERROR] An unexpected error occurred: {e}")
```
